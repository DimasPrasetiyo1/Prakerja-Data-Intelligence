```python
# Menemukan Korelasi Kolom
correlation_matrix = df.corr()
correlation_matrix
```

```python
# Melihat korelasi dengan Heatmap
fig, axes = plt.subplots(1,1,figsize=(9,6))
sns.heatmap(correlation_matrix, annot=True)
```

```python
plt.figure(figsize = (20,10))
sns.barplot(x=df['Sub-Category'], y=df['Profit'])
```

```python
# Mencari nilai profit dari masing masing sub kategori
df.groupby("State").Profit.agg(['sum', 'mean', 'count', 'min', 'max', 'median', 'std', 'var'])
```

```python
## Menampilkan visualisasi korelasi dengan Scatterplot antara Sales Dengan Profit
fig, axes = plt.subplots(1,1,figsize=(9,6))
axes.scatter(df['Sales'], df['Profit'])
axes.set_xlabel('Sales')
axes.set_ylabel('Profit')
```

```python
plt.figure(figsize = (10,10))
sns.displot(df['Sales'], bins=100, alpha=0.4)
plt.show()
```
